Here's a function that tells us actual rank given a row of the data frame above. This is just a prototype to figue out what's going wrong.

```python
def nonsumorank(row):
    counter = 1
    if df1983['rikishi1_rank'][row] == 'M'+str(counter)+'e':
        return counter
    if df1983['rikishi1_rank'][row] == 'M'+str(counter)+'w':
        return counter + 1
    else:
        return 0
```

This works fine. We can even test it on a given row.

```python
print(nonsumorank(10))
print(df1983['rikishi1_rank'][10])
```
